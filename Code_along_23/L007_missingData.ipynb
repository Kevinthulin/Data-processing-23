### Missing Data
- missing data occurs when values are simply absent or contain NaN (not a number) for any feature (column) in a given dataset. this will casue issues with many  machine learning algorithms.

**missing data can negativelty impact:**
- data visulizaion
- Arithemetic computation
- Machine learning algorithms

**Common methods to deal with missing data**
- remove rows or columns containing missing data
- impute with mean or median
- impute with mode (most frequently occuring feature)
- impute with forward or backward fill
- interpolate data between two points

*Note: Domain knowledge is often needed to decide how to fill nulls.*

```python
df = pd.DataFrame(np.random.randn(10, 4 ), columns=["A", "B", "C", "D"])
df = df[df > 0]
df
```

```python
copy = df.copy()
copy.drop(columns="D", inplace=True)
# copy.drop(index=7)
```

## dropna
removes rows (default), or columns, containing null

**parameters**
- **how** any or all
- **thresh** set numbers of Nan missing values in a row must contain to not drop
- **Subset** drops all NaN values in a single or multipale columns(rows)
- **axis** "index (default) or "columns"d

```python
copy.dropna(how="all")
copy.dropna(thresh=2)
copy.dropna(subset=["A", "B"])
copy.dropna(axis="columns", subset=2)
```
