### Update cells using loc or iloc
one or multiable cells can be set to a new value using loc or iloc
- **df.iloc[:] = 0** sets all *all* cells to 0
 

```python
import pandas as pd
```

```python
column_values = [f"Column{i}" for i in range(1, 5)]
index_values = [chr(i) for i in range(65, 72)]

df = pd.DataFrame([[f"{ind}{col}" for col in range(1,5)] for ind in index_values],
                  index=index_values, columns=column_values)
                

df
```

```python
copy = df.copy()
copy.iloc[1, 1] = "xxx"
copy.iloc[0, :] = "yy"
copy.loc[["C", "E", "G"], ["Column2", "Column4"]]= "ZZ"
copy
```

## Adding row or columns
roes and columns can be added using loc and not iloc

```python
copy = df.copy()
copy.loc["H"] = ["H1", "H2", "HTres", "H4"]
copy.loc[:, "Column5"] = copy.loc[:, "Column1"] + "-> "+ copy.loc[:, "Column1"]
copy["Column7"] = copy["Column2"] + "<->" + copy["Column4"]
copy.loc["I", ["Column3", "Column6"]] = 0
copy
```

# renameing colums and indexes

```python
copy = df.copy()
copy.columns= ["A", "B", "C", "D"]
copy.index = [1, 2, 3, 4, 5, 6, 7]
copy.rename(columns={"B": "Column B", "D": "Kevkos"}, inplace=True)
copy.rename(index={4: "hejdå"}, inplace=True)
copy
```

## Delete rows and columns

```python
copy = df.copy()
copy.drop(index=["D", "F", "G"], inplace=True)
copy
```
